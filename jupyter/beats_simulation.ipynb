```python
%matplotlib notebook
import matplotlib
matplotlib.use('nbagg')
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.arange(-300.0, 300.0, 0.01).reshape(-1, 1) # time grid
wstep = 1/(t[-1] - t[1]); w = 2 * np.pi * np.arange(0, len(t) * wstep, wstep) # frequency grid
```

```python
w1 = 0.95
w2 = 1
w3 = 1.15
T1 = T2 = T3 = 80
T2 = 20
```

```python
E1 = np.exp(-(t/T1)**2) * np.cos(w1*t)
E3 = np.exp(-(t/T3)**2) * np.cos(w3*t)
```

```python
delay = np.arange(-100, 100, 0.5).reshape(1,-1)
E2 = np.exp(-((t-delay)/T2)**2) * np.cos(w2*(t-delay))
```

```python
SF = (E1 + E2 + E3) ** 2
```

```python
SFspectrum = np.abs(np.fft.fft(SF, axis=0))**2
```

```python
plt.imshow(SFspectrum[130:250,:], aspect='auto', extent=(delay[0,0], delay[0,-1], w[249], w[130]))
```
